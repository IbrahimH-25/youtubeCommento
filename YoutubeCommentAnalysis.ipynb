# Importing dependancies

```python
from pandas import read_csv
from googleapiclient.discovery import build
from numpy import array,unique
```

# Looking at data

```python

data = r'youtoxic_english_1000.csv'
dataDF = read_csv(data)
dataDF
```

```python
catagoryList = dataDF.columns.values.tolist()[3:]
catagoryList
```

# Creating a model 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
import pickle as pk
```

```python
#vectorizing Column text
#starting with IsToxic
for colm in catagoryList:
    vecorModel = TfidfVectorizer(stop_words='english',max_features=1000)
    predictionGoal = colm
    x = dataDF['Text']
    y = dataDF[predictionGoal]
    print(dataDF[dataDF[predictionGoal] == True])

    vectorized_x = vecorModel.fit_transform(x)

    #making a randomforestcalssifier model
    rfc_model = RandomForestClassifier()
    rfc_model.fit(vectorized_x,y)
    rfcPkFile = pk.dump(rfc_model, open('pklFiles//%s_classifier.pkl'%predictionGoal,'wb'))
    vecPkFile = pk.dump(vecorModel,open('pklFiles//%s_vectorized_x.pkl'%predictionGoal,'wb'))
```

# Running model

```python
import pickle as pk

#creating a test data that will be the output
dataTestDF = dataDF[['CommentId','VideoId','Text']]
print(data)
# creating code for model


model = pk.load(open('yt_text_toxic_classifier.pkl','rb'))
tdif_vectorizer = pk.load(open('vectorized_x.pkl','rb'))
#lable_encoder = pk.load(open(''))

features = tdif_vectorizer.transform(dataTestDF['Text'])
predictions = model.predict(features)
#print(predictions)
dataTestDF[predictionGoal] = predictions
```
